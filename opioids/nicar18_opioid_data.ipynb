## Combining the two

Where the session demonstrated Excel's fantastic `vlookup`, I used pandas to merge the two tables.

```python
cdc_15_16 = pd.merge(cdc_2015, cdc_2016, how='outer')
```

```python
cdc_15_16['2015 Prescribing Rate'] = pd.to_numeric(cdc_15_16['2015 Prescribing Rate'], errors='coerce')
```

```python
cdc_15_16['2016 Prescribing Rate'] = pd.to_numeric(cdc_15_16['2016 Prescribing Rate'], errors='coerce')
```

```python
cdc_15_16.dropna(axis=0, how='any', inplace=True)
```

```python
cdc_15_16['diff'] = cdc_15_16['2016 Prescribing Rate']-cdc_15_16['2015 Prescribing Rate']
```

```python
cdc_15_16['pct_diff'] = (cdc_15_16['diff']/cdc_15_16['2015 Prescribing Rate'])*100
```

```python
cdc_15_16.sort_values(by='diff', ascending=False)
```

```python
cdc_15_16.to_csv('cdc_2015-2016.csv')
```
